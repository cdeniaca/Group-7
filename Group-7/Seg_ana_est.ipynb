# Gráficas varias

```python
andaluces = df_niv_est_pob[df_niv_est_pob['comunidad autonoma'] == 'Andalucia']
andaluces_sorted = andaluces.sort_values(by=['datos'], ascending=[True])

plt.figure(figsize=(10, 6))

plt.bar(andaluces['etapa educativa'], andaluces_sorted['datos'], color='grey')  
plt.title("Andaluces con una formacion concreta")
plt.xlabel("Etapa educativa")
plt.ylabel("Valores")
plt.xticks(rotation=45)
plt.show()
```

```python
catalunya = df_niv_est_pob[df_niv_est_pob['comunidad autonoma'] == 'Pais Vasco']
pv_sorted = andaluces.sort_values(by=['datos'], ascending=[True])

plt.figure(figsize=(10, 6))

plt.bar(catalunya['etapa educativa'], pv_sorted['datos'], color='grey')  
plt.title("Vascuences con una formacion concreta")
plt.xlabel("Etapa educativa")
plt.ylabel("Valores")
plt.xticks(rotation=45)
plt.show()
```

```python
top_5 = df_aband_est.groupby(['año', 'comunidad autonoma'])['total'].max().reset_index()
top_5_sorted = top_5.sort_values(by=[ 'total'], ascending=[True])
top_5_final = top_5_sorted.groupby('año').head(5)
año_concreto = 2018
top_5_año = top_5_final[top_5_final['año'] == año_concreto]

plt.figure(figsize=(10, 6))
plt.bar(top_5_año['comunidad autonoma'], top_5_año['total'], color='skyblue')

plt.xlabel('Comunidad autonoma')
plt.ylabel('porcentajes')
plt.title(f'Top 5 Comunidades Autónomas con mas abandonos de estudios por motivos laborales en {año_concreto}')
plt.xticks(rotation=45)
plt.show()
```
